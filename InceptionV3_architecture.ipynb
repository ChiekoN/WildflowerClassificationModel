# InceptionV3 architecture

I can see and visualize the InceptionV3 architecture.

![InceptionV3 architecture image](inceptionv3_architecture.png)



```python
%matplotlib notebook

import tensorflow as tf
from tensorflow import keras
```

```python
base_model = keras.applications.InceptionV3(include_top=False, weights='imagenet', input_shape=(224, 224, 3))
```

```python
base_model.summary()
```

```python
for i, layer in enumerate(base_model.layers):
    print("[{}] : {}".format(i, layer.name))
```

```python
keras.utils.plot_model(base_model, to_file="inceptionV3.png")
# To use plot_model(), you need to install:
#  - Graphviz : https://www.graphviz.org/
#  - Graphviz python interface: pip install GraphViz
#  - pydot: pip install pydot
```
